```python
def in_thousand (inst) :
    if inst == "500.0 k" :
        return "0.5 M"
    elif inst == "100.0 k" :
        return "0.1 M"
    else :
        return inst
```

```python
df['installs']= df['installs'].apply(in_thousand)
```

```python
df['installs']= df['installs'].str.replace( 'M', '').str.strip().astype('float')
```
